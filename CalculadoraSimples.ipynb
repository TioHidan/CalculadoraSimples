```python
print("Calculadora Simples")
print("-------------------\n \n")

num1= int (input("\nDigite o primeiro numero :"))
num2= int (input("\nDigite o segundo numero :"))

print("escolha uma operação : \n")
print("1. Adição (+)\n")
print("2. Subtração (-)\n")
print("3. Multiplicação (*)\n")
print("4. Divisão (/)\n")
opcao=int (input("Opção :"))

if opcao==1 :
  res=num1+num2
elif opcao==2 :
  res=num1-num2
elif opcao==3 :
  res=num1*num2
elif opcao==4 and num2!=0 :
  res=num1/num2
else :
  print("Opção invalida")


if opcao==1 :
  print("\nResultado",num1,"+",num2,"=",res)
elif opcao==2 :
  print("\nResultado",num1,"-",num2,"=",res)
elif opcao==3 :
  print("\nResultado",num1,"*",num2,"=",res)
elif opcao==4 and num2!=0 :
  print("\nResultado",num1,"/",num2,"=",res)
```
